```python
import random
import matplotlib.pyplot as plt

def generate_random_numbers(num_samples, lower_limit, upper_limit):
    return [random.randint(lower_limit, upper_limit) for _ in range(num_samples)]

if __name__ == "__main__":
    # Set the parameters
    num_samples = 10
    lower_limit = 1
    upper_limit = 10

    # Generate random numbers
    independent_variable = generate_random_numbers(num_samples, lower_limit, upper_limit)
    dependent_variable = generate_random_numbers(num_samples, lower_limit, upper_limit)

    # Plot histogram
    plt.scatter(independent_variable , dependent_variable)
    plt.xlabel('Independent Variable')
    plt.ylabel('Dependent Variable')
    plt.show()
```

```python
#CountPlot(B)
import seaborn as sns
sns.countplot( x = file['Education'] , data = file , hue = file['Loan_Status'])
plt.show()
```

```python
#Barplot(U)
gender_counts = file['Gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.show()
```
